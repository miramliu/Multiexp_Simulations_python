```python
b=b.T
data = df_AveragedNoise.iloc[j,2:].to_numpy().astype(float)
sumData2=sum(data**2)
# range for the parameters
logDgrid = 120
logD = np.linspace(-10,-4,logDgrid)
logDsgrid = 120
logDs = np.linspace(-7,0,logDsgrid)
fGrid = 60
f = np.linspace(0,1,fGrid)

logDmean = -6.2
logDstd = 1
logPrD = -0.5*(logD - logDmean)**2/logDstd**2 - 0.5*np.log(2*np.pi*logDstd**2)

logDsmean = -3.5#init: -3.5 
logDsStd = 1
logPrDs = -0.5*(logDs - logDsmean)**2/logDsStd**2 - 0.5*np.log(2*np.pi*logDsStd**2)


A = np.outer(np.exp(np.atleast_2d(logPrD).T.conj()),np.exp(logPrDs))
B=np.squeeze(np.tile(np.atleast_2d(logPrD).T.conj()[:,np.newaxis],(logDsgrid,1)))
Pr = A*B<np.tile(logDs[:,np.newaxis],(1,logDgrid))
logPr = np.log(Pr)

# various modelling arrays
#exp_b_logD = repmat(np.exp(-b*np.exp(logD)),[1 1 logDsgrid])
exp_b_logD = np.squeeze(np.tile(np.exp(-b[:,np.newaxis]*np.exp(logD))[:,np.newaxis,np.newaxis],(logDsgrid,1, 1)))
exp_b_logDs = np.squeeze(np.tile(np.exp(-b[:,np.newaxis]*np.exp(logDs))[:,np.newaxis,np.newaxis],[logDgrid,1, 1]))
exp_b_logDs = np.transpose(exp_b_logDs, (0, 2, 1))#.shape# permute(exp_b_logDs,[1, 3, 2])

# compute log-likelihood and log posterior by loonp.ping over f (loops in D
# and D* are incorporated here using array operations)
#[logLh,logPost] = deal(zeros(fGrid,logDgrid,logDsgrid))
logLh = np.zeros([fGrid,logDgrid,logDsgrid])
logPost=np.zeros([fGrid,logDgrid,logDsgrid])

for nf in range(1,fGrid):
    G = f[nf]*exp_b_logDs + (1-f[nf])*exp_b_logD
    denominator=sum(G**2)#*np.tile(weight[:,np.newaxis],[1, logDgrid, logDsgrid]))
    numerator = (sumData2 - sum(np.tile((data)[:,np.newaxis, np.newaxis],[1, logDgrid, logDsgrid])*G)**2)
    logLh[nf,:,:]=np.squeeze(-0.5*len(b)*np.log(numerator/denominator))
    #logLh(nf,:,:) = -0.5*length(b)*np.log((sumData2 - sum(np.tile((data*weight)[:,np.newaxis],[1, logDgrid, logDsgrid])*G)**2))/(sum(G**2*np.tile(weight[:,np.newaxis],[1, logDgrid, logDsgrid])))
    #logPost(nf,:,:) = logLh(nf,:,:) + reshape(logPr,[1 logDgrid logDsgrid]
    logPost[nf,:,:] = logLh[nf,:,:] + logPr.reshape(1, logDgrid, logDsgrid,order='F').copy()

# log-posterior
Post = np.exp(logPost - max(max(logPost)))
# normalisation
sP = sum(Post)
```
