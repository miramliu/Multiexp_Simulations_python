```python

#give name of error
def CalculatePE_fraction(df, trait):
    df[trait + ' PE'] = (df[trait]-df['true ' + trait])#*100/(df['true ' + trait])
    return df

def CalculatePE_diffusion(df, trait): #scale by 1000
    df[trait + ' PE'] = ((df[trait]/1000)-df['true ' + trait])#*100/(df['true ' + trait])
    return df
```

# First true data with no noise...

```python
df_TrueData = pd.merge(Truth, TrueDataFits, how = 'inner', left_on=["Run Number"], right_on = ['Run Number'])

df_TrueData = CalculatePE_fraction(df_TrueData, 'fast fraction')
df_TrueData = CalculatePE_fraction(df_TrueData, 'med fraction')
df_TrueData = CalculatePE_fraction(df_TrueData, 'slow fraction')
df_TrueData = CalculatePE_diffusion(df_TrueData, 'fast diffusion')
df_TrueData = CalculatePE_diffusion(df_TrueData, 'med diffusion')
df_TrueData = CalculatePE_diffusion(df_TrueData, 'slow diffusion')
```

```python
pl.hist(df_TrueData['fast fraction PE'], bins = 50, label = 'fast fraction fit difference')
pl.hist(df_TrueData['true fast fraction'], alpha = 0.4, label = 'distribution of true fast fraction')
pl.legend()
pl.show()

pl.hist(df_TrueData['med fraction PE'], bins = 50)
pl.hist(df_TrueData['true med fraction'], bins = 50, alpha = 0.4, label = 'distribution of true med fraction')
pl.legend()
pl.show()

pl.hist(df_TrueData['slow fraction PE'], bins = 50)
pl.hist(df_TrueData['true slow fraction'], bins = 50, alpha = 0.4, label = 'distribution of true slow fraction')
pl.legend()
pl.show()
```
