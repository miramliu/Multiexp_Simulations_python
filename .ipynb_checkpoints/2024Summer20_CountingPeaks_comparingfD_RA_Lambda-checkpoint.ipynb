```python
merged = pd.merge(RA_Database_lambdaCV[['Record ID','fD blood mean','fD tubule mean','fD tissue mean']],RA_Database_lambda8[['Record ID','fD blood mean', 'fD tubule mean','fD tissue mean']], how = 'inner', on = "Record ID")

merged = pd.merge(merged[['Record ID','fD blood mean_x','fD tubule mean_x','fD tissue mean_x','fD blood mean_y','fD tubule mean_y','fD tissue mean_y']],RA_Database_lambdapt1[['Record ID','fD blood mean', 'fD tubule mean','fD tissue mean']], how = 'inner', on = "Record ID")

def CCC(y_true, y_pred):
    """Concordance correlation coefficient."""
    # Raw data
    dct = {
        'y_true': y_true,
        'y_pred': y_pred
    }
    df = pd.DataFrame(dct)
    # Remove NaNs
    df = df.dropna()
    # Pearson product-moment correlation coefficients
    y_true = df['y_true']
    y_pred = df['y_pred']
    cor = np.corrcoef(y_true, y_pred)[0][1]
    # Means
    mean_true = np.mean(y_true)
    mean_pred = np.mean(y_pred)
    # Population variances
    var_true = np.var(y_true)
    var_pred = np.var(y_pred)
    # Population standard deviations
    sd_true = np.std(y_true)
    sd_pred = np.std(y_pred)
    # Calculate CCC
    numerator = 2 * cor * sd_true * sd_pred
    denominator = var_true + var_pred + (mean_true - mean_pred)**2

    return numerator / denominator
```

```python
pl.scatter(merged['fD blood mean_x'], merged['fD blood mean'])
pl.scatter(merged['fD tubule mean_x'], merged['fD tubule mean'])
pl.scatter(merged['fD tissue mean_x'], merged['fD tissue mean'])



AllCV = list(merged['fD blood mean_x']) + list(merged['fD tubule mean_x']) + list(merged['fD tissue mean_x'])
Allpt1 = list(merged['fD blood mean']) + list(merged['fD tubule mean']) + list(merged['fD tissue mean'])

print(CCC(AllCV, Allpt1))
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(AllCV, Allpt1)
r_value2 = r_value**2
x = np.linspace(0, max(AllCV))
pl.plot(x,func(x,slope,intercept),label = f'y= {slope:.2f}x +{intercept: .1f}, $R^2$={r_value2:.2f}',color = 'grey') 
pl.legend()
pl.title('CV vs Lambda pt1')
pl.show()
```

```python
pl.scatter(merged['fD blood mean_x'], merged['fD blood mean_y'])
pl.scatter(merged['fD tubule mean_x'], merged['fD tubule mean_y'])
pl.scatter(merged['fD tissue mean_x'], merged['fD tissue mean_y'])



AllCV = list(merged['fD blood mean_x']) + list(merged['fD tubule mean_x']) + list(merged['fD tissue mean_x'])
All8 = list(merged['fD blood mean_y']) + list(merged['fD tubule mean_y']) + list(merged['fD tissue mean_y'])

print(CCC(AllCV, All8))

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(AllCV, All8)
r_value2 = r_value**2
x = np.linspace(0, max(AllCV))
pl.plot(x,func(x,slope,intercept),label = f'y= {slope:.2f}x +{intercept: .1f}, $R^2$={r_value2:.2f}',color = 'grey') 
pl.legend()
pl.title('CV vs Lambda 8')
pl.show()
```
